```python
from bs4 import BeautifulSoup as bs
import requests
import re

import pandas as pd

import sqlalchemy as sal
from sqlalchemy import create_engine
from sqlalchemy.sql import text

from datetime import datetime
import time
import random
```

```python
years = range(2010, 2024)
```

```python
team_url_code_dict = {
    'Bills': 'BUF',
    'Ravens': 'RAV',
    'Texans': 'HTX',
    'Broncos': 'DEN',
    'Dolphins': 'MIA',
    'Bengals': 'CIN',
    'Colts': 'CLT',
    'Chiefs': 'KAN',
    'Patriots': 'NWE',
    'Browns': 'CLE',
    'Jaguars': 'JAX',
    'Raiders': 'RAI',
    'Jets': 'NYJ',
    'Steelers': 'PIT',
    'Titans': 'OTI',
    'Chargers': 'SDG',
    'Cowboys': 'DAL',
    'Bears': 'CHI',
    'Falcons': 'ATL',
    'Cardinals': 'CRD',
    'Giants': 'NYG',
    'Lions': 'DET',
    'Panthers': 'CAR',
    'Rams': 'RAM',
    'Eagles': 'PHI',
    'Packers': 'GNB',
    'Saints': 'NOR',
    '49ers': 'SFO',
    'Commanders': 'WAS',
    'Vikings': 'MIN',
    'Buccaneers': 'TAM',
    'Seahawks': 'SEA',
}
```

```python
def webscrape_head_coach (url:str):
    
    time.sleep(random.randrange(7,9))
    
    html = requests.get(url).text

    soup = bs(html, 'html.parser')

    top_banner_html = soup.find_all(class_='teams')[0]

    head_coaches = {'Head_Coach_Name': [], 'No_Games_Coached_per_Season': []}
    
    #check if p tag is coach then proceed if not, pass
    for p in top_banner_html.find_all('p'):
        
            #check if p tag is coach then proceed if not, pass
        for pc in p.contents:
            if 'Coach:' in pc:            
                for t in p.contents[2:]:
                    try:
                        coaches_name = t.contents[0]
                        
                    except:
                        split = t.split(' ')
                    
                        q = re.sub(r'[()]','', split[1])
                        games_list = q.split('-')
                        
                        games_list = list(map(int, games_list))
    
                        no_games = sum(games_list)
                        
                        for r in range(0,no_games):
                            head_coaches['Head_Coach_Name'].append(coaches_name)
                            head_coaches['No_Games_Coached_per_Season'].append(no_games)

    return pd.DataFrame(head_coaches)
```

```python
head_coach_df = pd.DataFrame()

for year in years:
    print(year, '***')
    for team in team_url_code_dict.keys():
        print(team)
        team_url = team_url_code_dict[team]
        
        url = f'https://www.pro-football-reference.com/teams/{team_url.lower()}/{year}.htm'

        df = webscrape_head_coach(url)
        df['Year'] = [year] * len(df)
        df['Team'] = [team] * len(df)
        df['Week'] = range(1, len(df)+1)
        
        head_coach_df = pd.concat([head_coach_df, df], axis=0)
```

```python
head_coach_df.reset_index(drop=True, inplace=True)
```

```python
head_coach_df.to_csv('Table Data/Head_Coaches')
```
