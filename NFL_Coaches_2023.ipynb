```python
from bs4 import BeautifulSoup as bs

import requests
import re


import pandas as pd
import numpy as np

import time

import random
```

```python
#find team links to each site
teams = [
    'azCardinals',
    'atlantaFalcons',
    'baltimoreRavens',
    'Buffalobills',
    'Panthers',
    'Bengals',
    'clevelandBrowns',
    'dallasCowboys',
    'denverBroncos',
    'detroitLions',
    'Packers',
    'houstonTexans',
    'Colts',
    'Jaguars',
    'Chiefs',
    'Raiders',
    'Chargers',
    'theRams',
    'miamiDolphins',
    'Vikings',
    'Patriots',
    'neworleansSaints',
    'Giants',
    'newyorkJets',
    'Steelers',
    '49ers',
    'Seahawks',
    'Buccaneers',
    'tennesseeTitans',
    'Commanders'
]
```

```python
teams = [team.lower() for team in teams]
teams
```

```python
str_url = 'http://www.{}.com/team/coaches-roster/'
```

```python
teams_url_li = [str_url.format(team) for team in teams]
teams_url_li.append('http://www.chicagobears.com/team/coaches/')
teams_url_li.append('http://www.philadelphiaeagles.com/team/coaches/')
```

```python
teams_url_li
```

```python
def coaches(teams_url_li: list):

    num_team = len(teams_url_li)

  #unable to join columns so appending list for now
  #coaches_df = pd.DataFrame()
    team_df_li = []
    
    coaches_col_li = []

    for url in range(0, num_team):

        html_request = requests.get(teams_url_li[url]).text
        time.sleep(random.randrange(2,4))

        html_coaches = bs(html_request, 'html.parser')
        
        
      #coaches names
    
        if url == 1:
            coaches_names = html_coaches.find_all('h3')[25:]
    
        else:
            coaches_names = html_coaches.find_all('h3')[:-3]

        num_coaches = len(coaches_names)


        coaches_li = [[coaches_names[i].contents[0][1:-1].strip() for i in range(0,num_coaches)]]


      #coaches title
        coaches_title = html_coaches.find_all('h5')

        num_coaches_title = len(coaches_title)

        coaches_title_li = [coaches_title[i].contents[0][1:-1].strip() for i in range(0, num_coaches_title)]

        data = [coaches_li[0][:num_coaches_title]]
        
        team_name = pd.DataFrame(
                        columns = [
                            'azCardinals',
                            'atlantaFalcons',
                            'baltimoreRavens',
                            'Buffalobills',
                            'Panthers',
                            'chicagoBears',
                            'Bengals',
                            'clevelandBrowns',
                            'dallasCowboys',
                            'denverBroncos',
                            'detroitLions',
                            'Packers',
                            'houstonTexans',
                            'Colts',
                            'Jaguars',
                            'Chiefs',
                            'Raiders',
                            'Chargers',
                            'theRams',
                            'miamiDolphins',
                            'Vikings',
                            'Patriots',
                            'neworleansSaints',
                            'Giants',
                            'newyorkJets',
                            'philapelphiaEagles',
                            'Steelers',
                            '49ers',
                            'Seahawks',
                            'Buccaneers',
                            'tennesseeTitans',
                            'Commanders'
                        ],
                        data=[[
                            'Cardinals',
                            'Falcons',
                            'Ravens',
                            'Buffalos',
                            'Panthers',
                            'Bears',
                            'Bengals',
                            'Browns',
                            'Cowboys',
                            'Broncos',
                            'Lions',
                            'Packers',
                            'Texans',
                            'Colts',
                            'Jaguars',
                            'Chiefs',
                            'Raiders',
                            'Chargers',
                            'Rams',
                            'Dolphins',
                            'Vikings',
                            'Patriots',
                            'Saints',
                            'Giants',
                            'Jets',
                            'Eagles',
                            'Steelers',
                            '49ers',
                            'Seahawks',
                            'Buccaneers',
                            'Titans',
                            'Commanders'
                        ]])

        df = pd.DataFrame(data=data, columns=coaches_title_li, index=[team_name.iloc[0,url]])

        team_df_li.append(df)

        
        for col in df.columns:
            coaches_col_li.append(col)

        coaches_df = pd.concat(team_df_li, ignore_index=True, axis=1)

    
    coaches_df.to_csv('Coaches_2023_df')

    coaches_df.columns = coaches_col_li
    
    return coaches_df
```

```python
coaches_df = coaches(teams_url_li)
```
