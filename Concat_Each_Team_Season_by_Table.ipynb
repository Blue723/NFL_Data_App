```python
#file name reference for each season
nfl_year_file_path = 'NFL_Data_{}'
```

```python
def file_path_df(file_path: None, years: list, teams:list, tables: list, columns: list):
            #prep for rows in df. 
    table_team_year_li = []
    for year in years:
        for team in teams:
            team_table_li = []
            table_team_year_li.append(team_table_li)
            for table in tables:
                team_table_li.append(nfl_year_file_path.format(year)+'\{}\\'.format(team)+table.format(team, year))
                
                
    
    #create dataframe with all year teams and tables
    table_names_team_year_df = pd.DataFrame(
        data=table_team_year_li, 
        columns=tables, 
        #index=team
    )
    
    
        #creating each file name in a df to use as reference when looping through open csv in the next step
    index_ranges = range(0, (len(table_team_year_li)+len(teams)), len(teams))
    table_name_by_year = []
    for i in index_ranges:
        tables_year_li = table_names_team_year_df.iloc[i-len(teams):i]
        table_name_by_year.append(tables_year_li)
    
    
    #pop blank table
    table_name_by_year.pop(0)
    
    
    return table_name_by_year
```

```python
def create_csv_each_season_tables (years: None, file_path_df: None, columns: list, team_names: list):
    #loop throuh each year and team's table 
    year_table = []
    for year,yi in zip(years,range(0, len(years))):
        for table_name in columns:
            
            table_df = pd.DataFrame()

            for i in file_path_df[yi][table_name]:
                df = pd.read_csv(i,index_col=0)
            
                table_df = pd.concat([table_df, df], axis=0, ignore_index=True)
            
            
            year_table.append(table_df)

            table_df.to_csv(nfl_year_file_path.format(year)+'\\'+ table_name.format('team',year))


    return year_table
```

```python
table_name_by_year = file_path_df(file_path=nfl_year_file_path, years=years, teams=team_names, tables=table_name_li, columns=table_col_name_li)
```

```python
year_table = create_csv_each_season_tables(years=years, file_path_df=table_name_by_year, columns=table_name_li, team_names=team_names)
```
