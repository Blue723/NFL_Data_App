```python
import pandas as pd

import os
```

```python
years = range(2018,2024)
```

```python
team_li=[
    '49ers',
    'Bears',
    'Bengals',
    'Broncos',
    'Browns',
    'Buccaneers',
    'Bills',
    'Cardinals',
    'Chargers',
    'Chiefs',
    'Colts',
    'Commanders',
    'Cowboys',
    'Dolphins',
    'Eagles',
    'Falcons',
    'Giants',
    'Jaguars',
    'Jets',
    'Lions',
    'Packers',
    'Panthers',
    'Patriots',
    'Raiders',
    'Rams',
    'Ravens',
    'Saints',
    'Seahawks',
    'Steelers',
    'Texans',
    'Titans',
    'Vikings'
]
```

```python
positions_df = pd.read_csv('Position Labels', index_col=0)
positions_df
```

```python
positions_df[positions_df['Pos'].str.contains('LA') == True]['Meaning'].values[0]
```

```python
# add position and name column to advanced logs
for year in years:
    for team in team_li:
        
        roster = pd.read_csv(f'NFL_Data_{year}/{team}/{team}_roster')

        player_file_li = os.listdir(f'NFL_Data_{year}/{team}/Players Log')
        
        for player_file in player_file_li:
            print(year, team, player_file)
            
            try:
                player = player_file.split('_')[0]
            except:
                player = player_file

            # filter roster with player name
            player_position_abbrev = roster[roster['Player'].str.contains(player) == True]['Position'].values[0]

            #use posisitons_df to index and get full name of position
            
            #handle double positions
            try:
                player_position_abbrev_li = player_position_abbrev.split('/')
                pos_li = []
                for abbrev in player_position_abbrev_li:
                    pos = positions_df[positions_df['Pos'].str.contains(abbrev) == True]['Meaning'].values[0]
                    pos_li.append(pos)
                
                player_position = '/'.join(pos_li)

            except:
                pass

            try:
                player_position = positions_df[positions_df['Pos'].str.contains(player_position_abbrev) == True]['Meaning'].values[0]
            except:
                player_position = player_position_abbrev

            player_file_loc = f'NFL_Data_{year}/{team}/Players Log/{player_file}'

            player_df = pd.read_csv(player_file_loc, index_col=0)

            #add team, position, player_name to datframe
            player_df['Team'] = [team] * len(player_df)
            player_df['Position'] = [player_position] * len(player_df)
            player_df['Player_Name'] = [player] * len(player_df)

            

            player_df.to_csv(player_file_loc)
```
