```python
filter_qb = qb_df[['Player', 'Position', 'Pass Completion/Attempts', 'Passing Yards/Attempts', 'Rushing Yards/Attempts',  
                   'Year', 'Week', 'Team', 'Team Points Scored', 'Win/Loss']]
filter_qb
```

```python
# filter for year

#### fix where query right now its taking out a lot of players that may only had been rushing or passing
# instead, i want to includeboth and leave out those that have neither stats

year = 2010





query = f'''
            with avg_rush as (
                select Player, Team, Year, "Rushing Attempts", "Rushing Yards"
                from qb_df 
                where "Rushing Attempts">0
            )
            select qb.Player, qb.Team, qb.Year,
            round(AVG(qb."Pass Completion/Attempts"),2) as "Average Pass Completion/Attempts",
            round(AVG(qb."Passing Yards/Attempts"),2) as "Average Passing Yards/Attempts",
            round(AVG(avg_rush."Rushing Yards"/avg_rush."Rushing Attempts"),2) as "Average Rushing Yards/Attempts"
            from qb_df qb
            full outer join avg_rush 
            on (qb.Player = avg_rush.Player)
            and (qb.Team = avg_rush.Team)
            and (qb.Year = avg_rush.Year)
            where "Pass Attempts" > 0
            group by qb.Player, qb.Team, qb.Year
'''

qb_pct = duckdb.sql(query).df()
qb_pct.columns
```

```python
qb_pct['Average Rushing Yards/Attempts'] = qb_pct['Average Rushing Yards/Attempts'].fillna(0)
```

```python
qb_pct_2010 = qb_pct[qb_pct['Year']==2010]
qb_pct_2010
```

```python
px.scatter_3d(qb_pct_2010, x='Average Pass Completion/Attempts', y='Average Passing Yards/Attempts', z='Average Rushing Yards/Attempts', hover_name='Team', color='Player')
```
