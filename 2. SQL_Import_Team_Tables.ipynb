```python
# union of all season and team to one table for each table
servername = '****'
football_data = 'Football_Data'
trusted_conneciton = '?trusted_conneciton=yes'
driver = '&driver=ODBC+Driver+17+for+SQL+Server'
```

```python
#create engine url to connect to sql server
engine_football_data = create_engine(f'mssql+pyodbc://@{servername}/{football_data}{trusted_conneciton}{driver}')
engine_football_data
```

```python
conn_engine_football_data = engine_football_data.connect()
```

```python
years = range(2010, 2024)
```

```python
team_li = [
     '49ers',
     'Bears',
     'Bengals',
     'Broncos',
     'Browns',
     'Buccaneers',
     'Bills',
     'Cardinals',
     'Chargers',
     'Chiefs',
     'Colts',
     'Commanders',
     'Cowboys',
     'Dolphins',
     'Eagles',
     'Falcons',
     'Giants',
     'Jaguars',
     'Jets',
     'Lions',
     'Packers',
     'Panthers',
     'Patriots',
     'Raiders',
     'Rams',
     'Ravens',
     'Saints',
     'Seahawks',
     'Steelers',
     'Texans',
     'Titans',
     'Vikings'
]
```

```python
table_li = [
    'team_stats_and_ranking',
    'schedule_and_game_results',
    'team_conversions',
    'passing',
    'rushing_and_receiving',
    'kick_and_punt_returns',
    'kicking',
    'punting',
    'defense_and_fumbles',
    'scoring_summary',
    'touchdown_log',
    'opponent_touchdown_log'
]
```

```python
#create table function that is imported to dataframe
def sql_create_table (file_name:str, connection: None, dataframe: None):

    dataframe.to_sql(file_name, con=connection, index=False, if_exists='replace')
    
    return None
```

```python
#import all data tables per year, team, table to sql db
for year in years:
    for team in team_li:
        for table in table_li:
            print(year, team, table)
            
            file = f'NFL_Data_{year}/{team}/{table}_{team}_{year}'
    
            #get dataframe
            df = pd.read_csv(file, index_col=0)
    
            #get file name for sql
            file_name = file.split(sep='/')[-1]

            #import file to sql database
            sql_create_table(file_name=file_name, connection=conn_engine_nfl_data, dataframe=df)
```
